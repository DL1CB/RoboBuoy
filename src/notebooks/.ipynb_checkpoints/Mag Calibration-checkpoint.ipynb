# Basic magnetometer calibration

```python
import time
import struct
import serial
import matplotlib.pyplot as plt


port = "COM4"

mx = []
my = []
mz = []

try:
    print("opening serial port")
    serialport = serial.Serial(port, 115200, timeout=0.1)
    time.sleep(3)
    serialport.readline()
    print(" collecting data ")
    for i in range(1000):
        line = serialport.readline()
        x,y,z = eval(line) 
        mx.append(x)
        my.append(y)
        mz.append(z)
        
except Error:
    pass
finally:
    serialport.close()

fig, ax = plt.subplots(1, 1)

ax.set_aspect(1)
ax.scatter(mx, my, color='r')
ax.scatter(my, mz, color='g')
ax.scatter(mz, mx, color='b')

minx = min(mx)
maxx = max(mx)
miny = min(my)
maxy = max(my)
minz = min(mz)
maxz = max(mz)
```

```python
print("X range: ", minx, maxx)
print("Y range: ", miny, maxy)
print("Z range: ", minz, maxz)

cx,cy,cz = ( (maxx + minx) / 2, (maxy + miny) / 2, (maxz + minz) / 2 )
print("Final calibration in uTesla:", cx,cy,cz)
```

```python
xc = [x - cx for x in mx]
yc = [y - cy for y in my]
zc = [z - cz for z in mz]
    
fig, ax = plt.subplots(1, 1)
ax.set_aspect(1)
ax.scatter(xc, yc, color='r')
ax.scatter(yc, zc, color='g')
ax.scatter(zc, xc, color='b')
```
